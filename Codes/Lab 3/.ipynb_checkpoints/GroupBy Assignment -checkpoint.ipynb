**GroupBy Assignment**                                    
1. Days when temperature was between 80 and 90                                
2. Days when it was between 50 and 60                                 
3. Days when it was anything else                           

```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\ratho\Desktop\5th Semester\ML\Codes\Lab 3\weather_by_cities.csv")
df
```

```python
g=df.groupby("city")
g
```

```python
for city,data in g:
    print("City : ",city)
    print("Data : ",data)
    print("\n")
```

```python
def grouping(temperature):
    if 80 <= temperature <= 90:
        return 'Group 1: 80-90'
    elif 50 <= temperature <= 60:
        return 'Group 2: 50-60'
    else:
        return 'Group 3: Other'
```

```python
df['group'] = df['temperature'].apply(grouping)
df
```

```python
groupedData = df.groupby('group')
```

```python
for group, data in groupedData:
    print(f"Group: {group}")
    print(data)
    print()
```
