```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\ratho\Desktop\5th Semester\ML\weather_data.csv")
df
```

```python
df.set_index('day',inplace=True)  #Setting the day as index
df
```

```python
new_df=df.fillna(0)
new_df
```

```python
new_df = df.fillna({    #Specifying individual columns about what should be written in place of empty places
    'temperature' : 0,
    'windspeed' : 0,
    'event' : 'No Event'
})
new_df
```

```python
new_df=df.fillna(method="ffill")  #It will fill the empty columns with value of the upper column
new_df       # method="bfill" will put below value
```

```python
new_df=df.fillna(method='ffill',limit=1)  #How many rows you want to fill i.e limit
new_df 
```

```python
new_df=df.fillna(method='ffill',axis=1)     
new_df    #Shift the values column wise, the value of left column will come to right if it is NaN
         #If done with bfill then values from right column will go to left one
```

```python
new_df=df.interpolate()      #Takes average
new_df
```

```python
new_df=df.dropna()
new_df
```

```python
new_df=df.dropna(how='all')  #Only delete if all null
new_df
```

```python
new_df=df.dropna(thresh=1)   #If even one record available then keeps the column
new_df
```

```python
df.dropna(thresh=0.3*len(df.columns))  #Keeps column if % data is avaiable else remove
df
```

```python
df.dropna(thresh=7,axis=1)
#Only keeps columns with minimum 7 values available
```
